```python
# Import all modules needed
import numpy as np  #Numeric python
import matplotlib.pyplot as plt  #Matlab style plotting
import scipy.integrate as sci # Integration toolbox

##Define some constant parameters
mass = 800/1000 # 800 grams


#Equations of motion
#F=m*a = m*zddot : second derivative of altitude as acceleration is second der of position and 1st der of speed
#Assuming flat earth model and so z axis is simply the straight altitude
# z is the altitude of the surface
# zdot is the velocity
# zddot is the acceleration

##Second order Differential equation

def Derivatives(state, t):
  ## Global variable
  global mass
  #state vectors - used to define the current position and velocity of the system
  z = state[0] #Position of spacecraft put in 1st row of state vector
  velz = state[1] #Velocity put in 2nd row of state vector
  # State vector [z
  #               velz]

  #Compute zdot - Kinematic relationships
  zdot = velz

  #Compute the total forces
  # Gravity
  gravity = -9.81*mass

  #Aerodynamics
  aero = 0.0

  # Thrust
  thrust = 0.0

  Forces = gravity + aero + thrust

  #Compute accelration
  zddot = Forces/mass

  #Now need to compute the statedot
  statedot = np.asarray([zdot, zddot])
  return statedot

```
